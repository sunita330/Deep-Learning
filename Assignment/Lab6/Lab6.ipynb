```python
import numpy as np

np.random.seed(42)

# =========================================================
# Step 1: Create Input Data
# =========================================================

# X represents a single feature varying from 0 to 10
# This range lets us observe early rapid growth and later slow growth

X = np.linspace(0,10,50).reshape(-1,1)


# =========================================================
# Step 2: Create Nonlinear Target
# =========================================================

# logarithmic relationship + small noise

noise = np.random.normal(0,0.2,(50,1))

y = np.log(X + 1) + noise

# This relationship is nonlinear.
# A straight line cannot capture both the rapid early growth
# and the later slow saturation.


# =========================================================
# Step 3: Initialize Model Parameters
# =========================================================

# input -> hidden layer (3 neurons)

W1 = np.random.uniform(-1,1,(1,3))
b1 = np.zeros((1,3))

# hidden -> output

W2 = np.random.uniform(-1,1,(3,1))
b2 = np.zeros((1,1))

# W1 controls how input contributes to hidden units
# W2 combines hidden features to produce final prediction


# =========================================================
# Step 4: Activation Function
# =========================================================

def activation(z):
    return np.maximum(0,z)      # ReLU

def activation_slope(z):
    return (z > 0).astype(float)

# ReLU keeps positive values and suppresses negative ones.
# Its slope tells the network how error should flow backward.


# =========================================================
# Step 5: Training Loop
# =========================================================

learning_rate = 0.01
epochs = 1000

for epoch in range(epochs):

    # ---------- Forward Pass ----------

    z1 = X @ W1 + b1
    h = activation(z1)

    y_hat = h @ W2 + b2


    # ---------- Error and Loss ----------

    error = y_hat - y

    loss = np.mean(error ** 2)


    # ---------- Backpropagation ----------

    # derivative of loss w.r.t prediction

    dL_dy = 2 * error / len(X)

    # gradients for output layer

    dL_dW2 = h.T @ dL_dy
    dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

    # propagate error to hidden layer

    dL_dh = dL_dy @ W2.T

    # apply activation slope

    dL_dz1 = dL_dh * activation_slope(z1)

    # gradients for first layer

    dL_dW1 = X.T @ dL_dz1
    dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)


    # ---------- Parameter Update ----------

    W1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1

    W2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2


    # ---------- Print Training Progress ----------

    if epoch % 100 == 0:
        print("Epoch:", epoch, "Loss:", loss)


# =========================================================
# Final Prediction
# =========================================================

y_pred = activation(X @ W1 + b1) @ W2 + b2

print("\nTraining finished.")
print("Final loss:", np.mean((y_pred - y)**2))
```
