```python
# ======================================================
# Step 1: Import Required Libraries
# ======================================================

import pandas as pd
import numpy as np


# ======================================================
# Step 2: Load the Dataset
# ======================================================

data = pd.read_csv("multiple_linear_regression_dataset.csv")

print("First 5 rows:\n")
print(data.head())

print("\nColumn Names:\n")
print(data.columns)

print("\nShape of dataset:\n")
print(data.shape)


# ======================================================
# Step 3: Separate Inputs and Output
# ======================================================

# Features (Inputs)
X = data[["age", "experience"]].values

# Target (Output
y = data["income"].values

print("\nShape of X:", X.shape)
print("Shape of y:", y.shape)


# ======================================================
# Step 4: Initialize Parameters
# ======================================================

n_features = X.shape[1]

# One weight per feature
w = np.zeros(n_features)

# Bias term
b = 0.0


# ======================================================
# Step 5: Define Forward Pass (Prediction)
# ======================================================

def predict(X, w, b):
    """
    Linear model:
    y_hat = w1*x1 + w2*x2 + b
    """
    return X.dot(w) + b


# ======================================================
# Step 6: Define Mean Squared Error (MSE)
# ======================================================

def mean_squared_error(y, y_hat):
    return ((y_hat - y) ** 2).mean()


# ======================================================
# Step 7: Compute Gradients
# ======================================================

def compute_gradients(X, y, y_hat):
    N = len(y)

    # Gradient w.r.t weights
    dw = (2 / N) * X.T.dot(y_hat - y)

    # Gradient w.r.t bias
    db = (2 / N) * (y_hat - y).sum()

    return dw, db


# ======================================================
# Step 8: Update Parameters
# ======================================================

def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b


# ======================================================
# Step 9: Training Loop
# ======================================================

lr = 0.0001
epochs = 1000

for epoch in range(epochs):

    # Forward pass
    y_hat = predict(X, w, b)

    # Compute loss
    loss = mean_squared_error(y, y_hat)

    # Compute gradients
    dw, db = compute_gradients(X, y, y_hat)

    # Update parameters
    w, b = update_parameters(w, b, dw, db, lr)

    # Print loss every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")


# ======================================================
# Step 10: Final Results
# ======================================================

print("\nFinal Weights:", w)
print("Final Bias:", b)


# ======================================================
# Step 11: Predict Salary for New Candidate
# ======================================================

new_candidate = np.array([4.5, 68])   # 4.5 years experience, test score 68
predicted_salary = new_candidate.dot(w) + b

print("\nPredicted Salary for new candidate:", predicted_salary)
```
